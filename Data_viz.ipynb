### Bar Plot: Top 10 Stocks by Average Trading Volume
#### Use this plot to compare the average trading activity of different stocks.

```python
# --- Bar Plot ---
# This aggregates the data to find the average volume per stock.
avg_volume = df.groupby('symbol')['volume'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
avg_volume.plot(kind='bar', color='skyblue')
plt.title('Top 10 Stocks by Average Trading Volume', fontsize=16)
plt.xlabel('Stock Symbol', fontsize=12)
plt.ylabel('Average Volume', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(os.path.join('stock_plots', 'top_10_stocks_by_volume_bar_chart.png'))
plt.show()
```

### Scatter Plot: High vs. Low Prices
#### This visualizes the relationship between two numerical variables.

```python
# --- Scatter Plot ---
plt.figure(figsize=(10, 8))
plt.scatter(aapl_df['High'], aapl_df['Low'], alpha=0.6, color='green')
plt.title('Relationship between High and Low Prices for AAPL', fontsize=16)
plt.xlabel('Daily High Price (USD)', fontsize=12)
plt.ylabel('Daily Low Price (USD)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join('stock_plots', 'aapl_high_low_scatter_plot.png'))
plt.show()
```
